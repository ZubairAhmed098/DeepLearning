```python
import os
from PIL import Image

# Method to resize images to 224 * 224
def resize_image(image, size):
    """Resize an image to the given size."""
    new_size = (size, size)
    return image.resize(new_size, Image.Resampling.LANCZOS)

# Method to loop over all the images in the given directory
def resize_images(input_dir, output_dir, size):
    """Resize the images in 'input_dir' and save into 'output_dir'."""
    for idir in os.scandir(input_dir):
        if not idir.is_dir():
            continue
        # Create output directory for this subfolder
        os.makedirs(os.path.join(output_dir, idir.name), exist_ok=True)
        images = os.listdir(idir.path)
        n_images = len(images)
        for iimage, image in enumerate(images):
            # Skipping non-image files
            if not image.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif', '.tiff')):
                continue
            try:
                image_path = os.path.join(idir.path, image)
                with Image.open(image_path) as img:
                    # Ensuring image is in RGB mode
                    img = img.convert('RGB')
                    img = resize_image(img, size)
                    img.save(os.path.join(output_dir, idir.name, image), img.format or 'JPEG')
            except (IOError, SyntaxError) as e:
                print(f"Error processing file {image}: {e}")
                continue
            # Logging progress every 1000 images
            if (iimage + 1) % 1000 == 0:
                print(f"[{iimage + 1}/{n_images}] Resized the images and saved into '{os.path.join(output_dir, idir.name)}'.")

input_dir = 'C:\\Users\\findz\\OneDrive\\Desktop\\practice\\python_codes\\Deep_learning_assignment\\original_images\\'
output_dir = 'C:\\Users\\findz\\OneDrive\\Desktop\\practice\\python_codes\\Deep_learning_assignment\\preprocessed_images\\'
size = 224
resize_images(input_dir, output_dir, size)
```

```python
""" Method to Create a vocabulary of unique question words and save it to a file."""
def make_q_vocab(src_dir, saving_dir):
    try:
        # Ensure directory exists
        os.makedirs(saving_dir, exist_ok=True)

        # Preparing regex and initialize vocabulary list
        regex = re.compile(r'(\W+)')
        q_vocab = []

        # Processing each .json file in the Questions folder
        dataset = [file for file in os.listdir(os.path.join(src_dir, 'Questions')) if file.endswith('.json')]
        for file in dataset:
            path = os.path.join(src_dir, 'Questions', file)
            print(f'processing: {path}')
            with open(path, 'r') as f:
                q_data = json.load(f)

            questions = q_data.get('questions', [])
            for quest in questions:
                split = regex.split(quest['question'].lower())
                tmp = [w.strip() for w in split if w.strip()]
                q_vocab.extend(tmp)

        # Deduplicate, sort, and add special tokens
        q_vocab = list(sorted(set(q_vocab)))
        q_vocab.insert(0, '<pad>')
        q_vocab.insert(1, '<unk>')

        # Save vocabulary to file
        with open(os.path.join(saving_dir, 'question_vocabs.txt'), 'w') as f:
            f.writelines([v + '\n' for v in q_vocab])

        print(f"Total unique words in question vocab: {len(q_vocab)}")

    except Exception as e:
        print(f"Error in make_q_vocab: {e}")
"""Function to Create a vocabulary of top answers and save it to a file."""
def make_a_vocab(src_dir, saving_dir, top_answer=1000):
    try:
        # Ensure directory exists
        os.makedirs(saving_dir, exist_ok=True)

        # Initialize answer frequency dictionary
        answers = defaultdict(lambda: 0)

        # Processing each .json file in the Annotations folder
        dataset = [file for file in os.listdir(os.path.join(src_dir, 'Annotations')) if file.endswith('.json')]
        for file in dataset:
            path = os.path.join(src_dir, 'Annotations', file)
            print(f'processing: {path}')
            with open(path, 'r') as f:
                data = json.load(f)

            annotations = data.get('annotations', [])
            for label in annotations:
                for ans in label.get('answers', []):
                    vocab = ans.get('answer', '').strip()
                    # Skip answers with non-alphanumeric characters
                    if not re.search(r'[^\w\s]', vocab):
                        answers[vocab] += 1

        # Sorting answers by frequency and keep top `top_answer`
        sorted_answers = sorted(answers.items(), key=lambda x: x[1], reverse=True)
        top_answers = ['<unk>'] + [ans for ans, _ in sorted_answers[:top_answer - 1]]

        # Saving top answers to file
        with open(os.path.join(saving_dir, 'annotation_vocabs.txt'), 'w') as f:
            f.writelines([ans + '\n' for ans in top_answers])

        print(f"Total unique answers: {len(answers)}")
        print(f"Saved top {top_answer} answers to vocab file.")

    except Exception as e:
        print(f"Error in make_a_vocab: {e}")
```

```python
# Function to tokenize a sentence
def tokenizer(sentence):
    # Compiling a regular expression to match non-word characters (i.e., punctuation and spaces)
    regex = re.compile(r'(\W+)')
    # Split the sentence by non-word characters, convert to lowercase, and remove any empty tokens
    tokens = regex.split(sentence.lower())
    # Strip any leading/trailing whitespace and filter out empty tokens
    tokens = [w.strip() for w in tokens if len(w.strip()) > 0]
    return tokens

# Function to match the top answers from an annotation answer list
def match_top_ans(annotation_ans):
    # Define the path to the annotation file
    annotation_dir = output_dir + '/annotation_vocabs.txt'

    # Check if 'top_ans' is not already initialized
    if "top_ans" not in match_top_ans.__dict__:
        # Read the annotation file and load top answers into a set
        with open(annotation_dir, 'r') as f:
            match_top_ans.top_ans = {line.strip() for line in f}

    # Extract the set of answers from the annotation answer list
    annotation_ans = {ans['answer'] for ans in annotation_ans}

    # Find the intersection between the loaded top answers and the annotation answers
    valid_ans = match_top_ans.top_ans & annotation_ans

    # If no valid answers are found, set the valid answers to a list containing '<unk>'
    if len(valid_ans) == 0:
        valid_ans = ['<unk>']
        # Increment the count of unknown answers
        match_top_ans.unk_ans += 1

    return annotation_ans, valid_ans
```

```python
# Function to preprocess the questions and annotations
def preprocessing(questions_json, annotation_dir, image_dir, labeled):
    filename = ''
    template = ''
    question_dict = {}
    questions = {}

    # Load the questions JSON file
    with open(questions_json, 'r') as f_question:
        data = json.load(f_question)
        questions = data['questions']
        # Determine the filename based on the dataset subtype (test-dev or other)
        if data['data_subtype'] == 'test-dev2015':
            filename = 'test2015'
        else:
            filename = data['data_subtype']  # Returns 'train2014' for training data
    print(filename)

    # If labeled data, load annotations
    if labeled:
        # Generating the file path pattern for annotation files based on the filename
        template = annotation_dir + f'/*{filename}*.json'
        annotation_path = glob.glob(template)[0]  # Get the first matching annotation file
        with open(annotation_path) as f:
            annotations = json.load(f)['annotations']

        # Create a dictionary mapping question_id to annotation answers
        question_dict = {ans['question_id']: ans for ans in annotations}

    # Initialize a counter for unknown answers
    match_top_ans.unk_ans = 0
    # Initialize dataset as a list with the same length as the questions list
    dataset = [None] * len(questions)
    # Process each question
    for idx, qu in enumerate(questions):
        if (idx+1) % 10000 == 0:
            print(f'Processing {data["data_subtype"]} data: {idx+1}/{len(questions)}')

        qu_id = qu['question_id']
        qu_sentence = qu['question']
        # Tokenize the question sentence
        qu_tokens = tokenizer(qu_sentence)
        img_id = qu['image_id']
        # Generating the image filename and path
        img_name = 'COCO_' + filename + '_{:0>12d}.jpg'.format(img_id)
        img_path = os.path.join(image_dir, img_name)

        # Store information about the question and image
        info = {'img_name': img_name,
                'img_path': img_path,
                'qu_sentence': qu_sentence,
                'qu_tokens': qu_tokens,
                'qu_id': qu_id}

        # If labeled, get the annotation answers and match top answers
        if labeled:
            annotation_ans = question_dict[qu_id]['answers']
            all_ans, valid_ans = match_top_ans(annotation_ans)
            info['all_ans'] = all_ans
            info['valid_ans'] = valid_ans

        # Store the question info in the dataset
        dataset[idx] = info

    # Print the total number of unknown answers encountered
    print(f'Total {match_top_ans.unk_ans} out of {len(questions)} answers are <unk>')
    return dataset
```

```python
# Function to process input data and save the processed dataset for model training
def process_data(image_dir, annotation_dir, question_dir, output_dir):
    processed_data = {}
    # Iterate over all files in the question directory
    for file in os.listdir(question_dir):
        # Extracting the datatype (train, val, etc.) from the file name
        datatype = file[20:-19]
        # Determine if the dataset is labeled (True for train/val, False for test)
        labeled = False if "test" in datatype else True
        # Getting the path to the questions JSON file
        questions_json = os.path.join(question_dir, file)
        # Preprocess the data for the current file and store it in the processed_data dictionary
        processed_data[datatype] = preprocessing(questions_json, annotation_dir, image_dir, labeled)

    # Combining 'train' and 'val' datasets into 'train-val'
    processed_data['train-val'] = processed_data['train'] + processed_data['val']

    # Saving each processed dataset as a .npy file
    for key, value in processed_data.items():
        # np.save to save the processed data into the output directory
        np.save(os.path.join(output_dir, f'{key}.npy'), np.array(value))
```

```python
INPUT_DIR = '/content/Mydrive/MyDrive/DeepLearning/Assignment/preprocess/'

class VQADataset(Dataset):
    def __init__(self, input_dir, input_file, max_ques_len = 30, transform = None, l_limit = None, u_limit = None):
        self.input_data = np.load(os.path.join(input_dir, input_file), allow_pickle=True)
        # Limiting the dataset for training if a limit is provided
        if l_limit is not None and u_limit is not None:
            self.input_data = self.input_data[l_limit:u_limit]
        # Loading vocabularies for questions and answers which are preprocessed
        self.ques_vocab = Vocabulary(os.path.join(input_dir, 'question_vocabs.txt'))
        self.ans_vocab = Vocabulary(os.path.join(input_dir, 'annotation_vocabs.txt'))
        #setting the max length of the question
        self.max_ques_len = max_ques_len
        self.labeled = True if not "test" in input_file else False
        #methods to perform some transformations on images
        self.transform = transform

    def __getitem__(self, idx):
        # Retrieve the image path for the current data sample
        path = self.input_data[idx]['img_path']
        # Loading the image from the file, convert it to RGB format, and transform it into a NumPy array
        img = np.array(Image.open(path).convert('RGB'))
        # Getting the question ID for the current data sample
        qu_id = self.input_data[idx]['qu_id']
        # Retrieve the tokenized question (a list of word tokens)
        qu_tokens = self.input_data[idx]['qu_tokens']
        # Initializing a fixed-length array for the question, filled with <pad> token indices
        qu2idx = np.array([self.ques_vocab.word2idx('<pad>')] * self.max_ques_len)
        # Replace the beginning of the array with indices of the actual question tokens
        # Truncates the tokens if they exceed the max question length
        qu2idx[:len(qu_tokens)] = [self.ques_vocab.word2idx(token) for token in qu_tokens]
        # Create the initial sample dictionary with the image, question, and question ID
        sample = {
            'image': img,
            'question': qu2idx,
            'question_id': qu_id
        }

        # If the dataset is labeled (not as test set), process the answer data
        if self.labeled:
            # Converting all valid answers for the current sample into their respective indices
            ans2idx = [self.ans_vocab.word2idx(ans) for ans in self.input_data[idx]['valid_ans']]
            # Randomly selecting one answer index from the valid answers
            ans2idx = np.random.choice(ans2idx)
            # populate the selected answer index to the sample
            sample['answer'] = ans2idx
        # If a transformation function is provided, apply it to the image
        if self.transform:
            sample['image'] = self.transform(sample['image'])
        # Return the final processed sample
        return sample
    #Method to give length of dataset
    def __len__(self):
        return len(self.input_data)
```

```python
#Creating dataloader function to create the train and test dataloaders
def data_loader(input_dir, batch_size, max_ques_len, num_worker, train_val_limits):
    transform = transforms.Compose([
        transforms.ToTensor(),  # convert to (C,H,W) and [0,1]
        transforms.Normalize((0.485, 0.456, 0.406), (0.229, 0.224, 0.225))  # mean=0; std=1
    ])
    # Creating upper and lower limits to questions as the total no of question are huge.
    if len(train_val_limits) == 4:
        tl_limit = train_val_limits[0]
        tu_limit = train_val_limits[1]
        vl_limit = train_val_limits[2]
        vu_limit = train_val_limits[3]
    else:
        tl_limit = None
        tu_limit = None
        vl_limit = None
        vu_limit = None
    #Creating VQA datset objects
    vqa_dataset = {
        'train': VQADataset(input_dir=input_dir, input_file='train.npy', max_ques_len=max_ques_len,
            transform=transform, l_limit=tl_limit, u_limit=tu_limit), #setting the upper and lower limit to check the training process
        'val': VQADataset(input_dir=input_dir, input_file='val.npy', max_ques_len=max_ques_len,
            transform=transform, l_limit=vl_limit, u_limit=vu_limit)
    }
    #Creating dataloader objects for train & valdation
    dataloader = {
        key: DataLoader(dataset=vqa_dataset[key], batch_size=batch_size,
            shuffle=True, num_workers=num_worker) for key in ['train', 'val']
    }

    return dataloader
```

```python
#Creating vocabulary Class to store all vocab info
class Vocabulary:
    def __init__(self, vocabulary_file_path):
        #Initialize the Vocabulary object.
        # Load vocabulary from the provided file path
        self.vocabulary = self._load_vocabulary(vocabulary_file_path)
        # Create a mapping from words to indices
        self.vocabulary2idx = {word: idx for idx, word in enumerate(self.vocabulary)}
        # Store the total size of the vocabulary
        self.vocabulary_size = len(self.vocabulary)

    def _load_vocabulary(self, vocabulary_file_path):
        #Load vocabulary from a file.
        with open(vocabulary_file_path, 'r') as file:
            # Read each line, strip extra whitespace, and return as a list
            vocabulary = [line.strip() for line in file]
        return vocabulary

    def word2idx(self, word):
        #Convert a word to its corresponding index.
        # Return the index of the word or the index of '<unk>' if the word is not in the vocabulary
        return self.vocabulary2idx.get(word, self.vocabulary2idx.get('<unk>'))

    def idx2word(self, idx):
        #Convert an index back to its corresponding word.
        return self.vocabulary[idx]
```

```python
class ImageEncoder(nn.Module):
    def __init__(self, embed_dim):
        super(ImageEncoder, self).__init__()
        # Loading a pretrained VGG19 model
        self.model = models.vgg19(pretrained=True)
        # Geting the number of input features for the last fully connected layer
        in_features = self.model.classifier[-1].in_features
        # Removing the last layer of VGG19's classifier (the final fully connected layer for classification)
        self.model.classifier = nn.Sequential(*list(self.model.classifier.children())[:-1])
        # Adding a new fully connected layer to map features to the desired embedding dimension
        self.fc = nn.Linear(in_features, embed_dim)

    def forward(self, image):
        # Extracting features of the image using the modified VGG19 model
        with torch.no_grad():  # Freezing the weights of the pretrained model during this pass
            img_feature = self.model(image)  # Output shape: (batch_size, feature_dim)
        #features to the embedding dimension
        img_feature = self.fc(img_feature)  # Output shape: (batch_size, embed_dim)
        # Applying L2 normalization to the features for better similarity comparisons
        l2_norm = F.normalize(img_feature, p=2, dim=1).detach()  # Normalize along the feature dimension

        return l2_norm
```

```python
class QuesEncoder(nn.Module):
    def __init__(self, ques_vocab_size, word_embed, hidden_size, num_hidden, qu_feature_size):
        super(QuesEncoder, self).__init__()
        # Embedding layer to map question words to word embeddings
        self.word_embedding = nn.Embedding(ques_vocab_size, word_embed)
        # Activation function to add non-linearity to embeddings
        self.tanh = nn.Tanh()
        # LSTM layer for sequential processing of question embeddings
        # Takes word embeddings as input and outputs hidden states
        self.lstm = nn.LSTM(word_embed, hidden_size, num_hidden)  # (input_dim, hidden_dim, num_layers)
        # Fully connected layer to transform the concatenated LSTM states to the desired feature size
        self.fc = nn.Linear(2 * num_hidden * hidden_size, qu_feature_size)

    def forward(self, question):
        # Map question words to embeddings
        ques_embedding = self.word_embedding(question) # Shape: (batch_size, question_length, word_embed)
        ques_embedding = self.tanh(ques_embedding) # Applying Tanh activation to the embeddings
        ques_embedding = ques_embedding.transpose(0, 1) # Transpose for LSTM input: (question_length, batch_size, word_embed)
        # Passing embeddings through the LSTM
        # Outputs: LSTM outputs (_) and final hidden states (hidden, cell)
        # hidden and cell shapes: (num_layers, batch_size, hidden_size)
        _, (hidden, cell) = self.lstm(ques_embedding)
        # Concatenating the hidden and cell states along the feature dimension
        # Shape: (num_layers, batch_size, 2 * hidden_size)
        ques_feature = torch.cat((hidden, cell), dim=2)
        ques_feature = ques_feature.transpose(0, 1) # Transpose for batch-first format: (batch_size, num_layers, 2 * hidden_size)
        ques_feature = ques_feature.reshape(ques_feature.size(0), -1) # Flattening the feature tensor:(batch_size, num_layers * 2 * hidden_size)
        ques_feature = self.tanh(ques_feature) # Applying Tanh activation to the flattened features
        # Transforming the features to the desired output size: (batch_size, qu_feature_size)
        ques_feature = self.fc(ques_feature)

        return ques_feature
```

```python
class VQAModel(nn.Module):
    def __init__(self, feature_size, ques_vocab_size, ans_vocab_size, word_embed, hidden_size, num_hidden):
        super(VQAModel, self).__init__()
        # Encoder to extract image features
        self.img_encoder = ImageEncoder(feature_size)
        # Encoder to extract question features
        self.ques_encoder = QuesEncoder(ques_vocab_size, word_embed, hidden_size, num_hidden, feature_size)
        # Dropout layer to prevent overfitting
        self.dropout = nn.Dropout(0.5)
        # Tanh activation function for non-linearity
        self.tanh = nn.Tanh()
        # Fully connected layer to map combined features to answer space
        self.fc1 = nn.Linear(feature_size, ans_vocab_size)
        # Second fully connected layer to refine logits in the answer space
        self.fc2 = nn.Linear(ans_vocab_size, ans_vocab_size)

    def forward(self, image, question):
        # Extract image features using the image encoder
        img_feature = self.img_encoder(image) # Output shape: (batch_size, feature_size)

        # Extract question features using the question encoder
        qst_feature = self.ques_encoder(question) # Output shape: (batch_size, feature_size)

        # Combine image and question features element-wise (Hadamard product)
        combined_feature = img_feature * qst_feature # Output shape: (batch_size, feature_size)

        combined_feature = self.dropout(combined_feature) # Apply dropout for regularization
        combined_feature = self.tanh(combined_feature)

        # Map combined features to the answer space using the first fully connected layer
        combined_feature = self.fc1(combined_feature) # Output shape: (batch_size, ans_vocab_size)

        # Apply another round of dropout for regularization
        combined_feature = self.dropout(combined_feature)
        combined_feature = self.tanh(combined_feature)

        # Refine logits using the second fully connected layer
        logits = self.fc2(combined_feature) # Output shape: (batch_size, ans_vocab_size)

        return logits
```

```python
# Function to train the VQA model
def train_VQA_model():
    # Initialize data loader
    train_val_limits = [0, 50000, 0, 10000]
    dataloader = data_loader(input_dir=DATA_DIR, batch_size=BATCH_SIZE, max_ques_len=MAX_QU_LEN, num_worker=NUM_WORKER, train_val_limits=train_val_limits)
    qu_vocab_size = dataloader['train'].dataset.ques_vocab.vocabulary_size  # Question vocabulary size
    ans_vocab_size = dataloader["train"].dataset.ans_vocab.vocabulary_size  # Answer vocabulary size

    # Initialize the VQA model
    model = VQAModel(feature_size=FEATURE_SIZE, ques_vocab_size=qu_vocab_size, ans_vocab_size=ans_vocab_size,
                     word_embed=WORD_EMBED, hidden_size=HIDDEN_SIZE, num_hidden=NUM_HIDDEN).to(device)

    # Set up optimizer, learning rate scheduler, and loss function
    optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE)
    scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=STEP_SIZE, gamma=GAMMA)
    criterion = nn.CrossEntropyLoss()

    print('<---- start training ---->')
    start_time = time.time()

    # Training loop over epochs
    for epoch in range(EPOCH):
        epoch_loss = {key: 0 for key in ['train', 'val']}  # Initialize loss for train/val

        model.train()  # Setting model to training mode
        for idx, sample in enumerate(dataloader['train']):
            # Adding prints to see how many batches are finished.
            if (idx + 1) % 1000 == 0:
                print(f'processing train data: {idx + 1}/{len(dataloader["train"])}')

            # Move data to device (GPU/CPU)
            image = sample['image'].to(device=device)
            question = sample['question'].to(device=device)
            label = sample['answer'].to(device=device)

            # Forward pass
            logits = model(image, question)
            loss = criterion(logits, label)

            epoch_loss['train'] += loss.item()  # Accumulating training loss

            # Backward pass
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

        # Validation loop
        model.eval()  # Setting model to evaluation mode
        for idx, sample in enumerate(dataloader['val']):
            if (idx + 1) % 1000 == 0:
                print(f'processing val data: {idx + 1}/{len(dataloader["val"])}')

            # Move data to device
            image = sample['image'].to(device=device)
            question = sample['question'].to(device=device)
            label = sample['answer'].to(device=device)

            # Disable gradient calculation
            with torch.no_grad():
                logits = model(image, question)
                loss = criterion(logits, label)

            epoch_loss['val'] += loss.item()  # Accumulate validation loss

        # Calculate average loss for training and validation
        for phase in ['train', 'val']:
            epoch_loss[phase] /= len(dataloader[phase])

            # Save logs to file
            with open(os.path.join(LOG_DIR, f'{phase}_log.txt'), 'a') as f:
                f.write(f'{epoch + 1}\t{epoch_loss[phase]}\n')

        # Print progress after each epoch
        print(f'Epoch:{epoch + 1}/{EPOCH} | Training Loss: {epoch_loss["train"]:.6f} | Validation Loss: {epoch_loss["val"]:.6f}')

        # Step the learning rate scheduler
        scheduler.step()

        # Check for early stopping
        early_stop = early_stopping(model, epoch_loss['val'])

        # Save model checkpoint every 5 epochs
        if (epoch + 1) % 10 == 0:
            torch.save(model.state_dict(), os.path.join(CKPT_DIR, f'model-epoch-{epoch + 1}.pth'))

        # If early stopping condition met, break the loop
        if early_stop:
            print(f'>> Early stop at epoch {epoch + 1}')
            break

    end_time = time.time()
    training_time = end_time - start_time
    print(f">> Finishing training | Training Time: {training_time // 60:.0f}m:{training_time % 60:.0f}s")


# Adding Early stopping mechanism
def early_stopping(model, epoch_loss, patience=7):
    early_stop = False
    # Initialize early stopping tracking variables
    if not bool(early_stopping.__dict__):
        early_stopping.best_loss = epoch_loss
        early_stopping.record_loss = epoch_loss
        early_stopping.counter = 0

    # If current loss is better than the best loss, save model
    if epoch_loss < early_stopping.best_loss:
        early_stopping.best_loss = epoch_loss
        torch.save(model.state_dict(), os.path.join(CKPT_DIR, 'best_model.pth'))

    # If loss starts increasing, track counter
    if epoch_loss > early_stopping.record_loss:
        early_stopping.counter += 1
        if early_stopping.counter >= patience:  # If counter exceeds patience value, stop early
            early_stop = True
    else:
        early_stopping.counter = 0
        early_stopping.record_loss = epoch_loss

    return early_stop
```

```python
class QuesEncoder(nn.Module):
    def __init__(self, ques_vocab_size, word_embed, hidden_size, num_hidden, qu_feature_size):
        super(QuesEncoder, self).__init__()
        # Embedding layer to map question words to word embeddings
        self.word_embedding = nn.Embedding(ques_vocab_size, word_embed)
        # Activation function to add non-linearity to embeddings
        self.tanh = nn.Tanh()
        # LSTM layer for sequential processing of question embeddings
        # Takes word embeddings as input and outputs hidden states
        self.lstm = nn.LSTM(word_embed, hidden_size, num_hidden)  # (input_dim, hidden_dim, num_layers)
        # Fully connected layer to transform the concatenated LSTM states to the desired feature size
        self.fc = nn.Linear(2 * num_hidden * hidden_size, qu_feature_size)

    def forward(self, question):
        # Map question words to embeddings
        ques_embedding = self.word_embedding(question) # Shape: (batch_size, question_length, word_embed)
        ques_embedding = self.tanh(ques_embedding) # Applying Tanh activation to the embeddings
        ques_embedding = ques_embedding.transpose(0, 1) # Transpose for LSTM input: (question_length, batch_size, word_embed)
        # Passing embeddings through the LSTM
        # Outputs: LSTM outputs (_) and final hidden states (hidden, cell)
        # hidden and cell shapes: (num_layers, batch_size, hidden_size)
        _, (hidden, cell) = self.lstm(ques_embedding)
        # Concatenating the hidden and cell states along the feature dimension
        # Shape: (num_layers, batch_size, 2 * hidden_size)
        ques_feature = torch.cat((hidden, cell), dim=2)
        ques_feature = ques_feature.transpose(0, 1) # Transpose for batch-first format: (batch_size, num_layers, 2 * hidden_size)
        ques_feature = ques_feature.reshape(ques_feature.size(0), -1) # Flattening the feature tensor:(batch_size, num_layers * 2 * hidden_size)
        ques_feature = self.tanh(ques_feature) # Applying Tanh activation to the flattened features
        # Transforming the features to the desired output size: (batch_size, qu_feature_size)
        ques_feature = self.fc(ques_feature)

        return ques_feature

class ImageEncoder_attn(nn.Module):
    def __init__(self, embed_dim):
        super(ImageEncoder_attn, self).__init__()
        # Load a pretrained VGG19 model
        self.model = models.vgg19(pretrained=True).features
        # Adding a 1x1 convolutional layer to map features to the desired embedding dimension
        self.conv = nn.Conv2d(512, embed_dim, kernel_size=1)

    def forward(self, image):
        # Extracting spatial features of the image using the modified VGG19 model
        with torch.no_grad():  # Freezing the weights of the pretrained model during this pass
            img_features = self.model(image)  # Shape: (batch_size, 512, H, W)

        # Map features to the desired embedding dimension
        img_features = self.conv(img_features)  # Shape: (batch_size, embed_dim, H, W)

        # Flatten spatial dimensions to get per-region features
        img_features = img_features.flatten(2).permute(0, 2, 1)  # Shape: (batch_size, num_regions, embed_dim)

        return img_features


class VQAModel_attn(nn.Module):
    def __init__(self, feature_size, ques_vocab_size, ans_vocab_size, word_embed, hidden_size, num_hidden):
        super(VQAModel_attn, self).__init__()

        # Encoder to extract image features
        self.img_encoder = ImageEncoder_attn(feature_size)

        # Encoder to extract question features
        self.ques_encoder = QuesEncoder(ques_vocab_size, word_embed, hidden_size, num_hidden, feature_size)

        # Attention mechanism layers
        self.attention_fc = nn.Linear(2 * feature_size, 1)  # For compatibility scoring

        # Dropout layer
        self.dropout = nn.Dropout(0.5)

        # Fully connected layers for answer prediction
        self.fc1 = nn.Linear(feature_size, ans_vocab_size)
        self.fc2 = nn.Linear(ans_vocab_size, ans_vocab_size)

    def forward(self, image, question):
        # Extract image features (batch_size, num_regions, feature_size)
        img_features = self.img_encoder(image)

        # Extract question features (batch_size, feature_size)
        qst_feature = self.ques_encoder(question)

        # Ensure qst_feature has the correct dimensions
        # Expand to (batch_size, 1, feature_size), then repeat to match num_regions
        qst_feature_exp = qst_feature.unsqueeze(1).expand(-1, img_features.size(1), -1)

        #print(f"img_features shape: {img_features.shape}")
        #print(f"qst_feature shape: {qst_feature.shape}")
        #print(f"qst_feature_exp shape: {qst_feature_exp.shape}")
        # Concatenate image and question features along the last dimension
        # Shape: (batch_size, num_regions, 2 * feature_size)
        combined_features = torch.cat([img_features, qst_feature_exp], dim=-1)

        # Compute attention scores for each region
        # Shape: (batch_size, num_regions, 1)
        attention_scores = self.attention_fc(combined_features)

        # Apply softmax to get attention weights
        # Shape: (batch_size, num_regions)
        attention_weights = F.softmax(attention_scores.squeeze(-1), dim=1)

        # Compute the weighted sum of image features
        # Shape: (batch_size, feature_size)
        attended_img_feature = torch.sum(img_features * attention_weights.unsqueeze(-1), dim=1)

        # Combine attended image features with question features
        combined_feature = attended_img_feature + qst_feature

        # Dropout and fully connected layers for answer prediction
        combined_feature = self.dropout(combined_feature)
        combined_feature = F.relu(self.fc1(combined_feature))
        logits = self.fc2(combined_feature)

        return logits

```

```python
# Function to train the VQA model
def train_VQA_model_attn():
    # Initialize data loader
    train_val_limits = [0, 50000, 0, 10000]
    dataloader = data_loader(input_dir=DATA_DIR, batch_size=BATCH_SIZE, max_ques_len=MAX_QU_LEN, num_worker=NUM_WORKER, train_val_limits=train_val_limits)
    qu_vocab_size = dataloader['train'].dataset.ques_vocab.vocabulary_size  # Question vocabulary size
    ans_vocab_size = dataloader["train"].dataset.ans_vocab.vocabulary_size  # Answer vocabulary size

    # Initialize the VQA model
    model = VQAModel_attn(feature_size=FEATURE_SIZE, ques_vocab_size=qu_vocab_size, ans_vocab_size=ans_vocab_size,
                     word_embed=WORD_EMBED, hidden_size=HIDDEN_SIZE, num_hidden=NUM_HIDDEN).to(device)

    # Set up optimizer, learning rate scheduler, and loss function
    optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE)
    scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=STEP_SIZE, gamma=GAMMA)
    criterion = nn.CrossEntropyLoss()

    print('<---- start training ---->')
    start_time = time.time()

    # Training loop over epochs
    for epoch in range(EPOCH):
        epoch_loss = {key: 0 for key in ['train', 'val']}  # Initialize loss for train/val

        model.train()  # Setting model to training mode
        for idx, sample in enumerate(dataloader['train']):
            # Adding prints to see how many batches are finished.
            if (idx + 1) % 1000 == 0:
                print(f'processing train data: {idx + 1}/{len(dataloader["train"])}')

            # Move data to device (GPU/CPU)
            image = sample['image'].to(device=device)
            question = sample['question'].to(device=device)
            label = sample['answer'].to(device=device)

            # Forward pass
            logits = model(image, question)
            loss = criterion(logits, label)

            epoch_loss['train'] += loss.item()  # Accumulating training loss

            # Backward pass
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

        # Validation loop
        model.eval()  # Setting model to evaluation mode
        for idx, sample in enumerate(dataloader['val']):
            if (idx + 1) % 1000 == 0:
                print(f'processing val data: {idx + 1}/{len(dataloader["val"])}')

            # Move data to device
            image = sample['image'].to(device=device)
            question = sample['question'].to(device=device)
            label = sample['answer'].to(device=device)

            # Disable gradient calculation
            with torch.no_grad():
                logits = model(image, question)
                loss = criterion(logits, label)

            epoch_loss['val'] += loss.item()  # Accumulate validation loss

        # Calculate average loss for training and validation
        for phase in ['train', 'val']:
            epoch_loss[phase] /= len(dataloader[phase])

            # Save logs to file
            with open(os.path.join(LOG_DIR, f'{phase}_attn_log.txt'), 'a') as f:
                f.write(f'{epoch + 1}\t{epoch_loss[phase]}\n')

        # Print progress after each epoch
        print(f'Epoch:{epoch + 1}/{EPOCH} | Training Loss: {epoch_loss["train"]:.6f} | Validation Loss: {epoch_loss["val"]:.6f}')

        # Step the learning rate scheduler
        scheduler.step()

        # Check for early stopping
        early_stop = early_stopping(model, epoch_loss['val'])

        # Save model checkpoint every 5 epochs
        if (epoch + 1) % 10 == 0:
            torch.save(model.state_dict(), os.path.join(CKPT_DIR, f'model-attn-epoch-{epoch + 1}.pth'))

        # If early stopping condition met, break the loop
        if early_stop:
            print(f'>> Early stop at epoch {epoch + 1}')
            break

    end_time = time.time()
    training_time = end_time - start_time
    print(f">> Finishing training | Training Time: {training_time // 60:.0f}m:{training_time % 60:.0f}s")


# Adding Early stopping mechanism
def early_stopping(model, epoch_loss, patience=7):
    early_stop = False
    # Initialize early stopping tracking variables
    if not bool(early_stopping.__dict__):
        early_stopping.best_loss = epoch_loss
        early_stopping.record_loss = epoch_loss
        early_stopping.counter = 0

    # If current loss is better than the best loss, save model
    if epoch_loss < early_stopping.best_loss:
        early_stopping.best_loss = epoch_loss
        torch.save(model.state_dict(), os.path.join(CKPT_DIR, 'best_model.pth'))

    # If loss starts increasing, track counter
    if epoch_loss > early_stopping.record_loss:
        early_stopping.counter += 1
        if early_stopping.counter >= patience:  # If counter exceeds patience value, stop early
            early_stop = True
    else:
        early_stopping.counter = 0
        early_stopping.record_loss = epoch_loss

    return early_stop
```
